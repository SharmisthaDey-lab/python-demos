##**Cab Driver-Rider Sharing Demo**

This code snippet used object oriended programing concepts to build relations ships between Driver & Rider parent classes and reflects various OOPs pillars like inheritence, encapsulation, polymorphism, abstractions at various stages in the code. This is a avery basic demo code from one of my hands-on activity.


##Step 1. Introducing Driver Class

```python
class Driver:
  def __init__(self,name,vehical,rating =5.0):
    self.name = name #Public attributes
    self.vehical = vehical #Public attribute
    self._rating = rating #Private attribute
    self.is_available = True

#Encapsulation to access protected variables using getter & setter methods
#Getter for protected attribute
  def get_rating(self):
    return self._rating
#Setter for proteccted attribute given by user
  def set_rating(self,new_rating):
    if 0 <= new_rating <= 5.0: # Corrected condition
      self._rating = new_rating
    else:
      print("Invalid rating")
#Polymorphism with __str__ -defines how drives is printed
  def __str__(self):
    return f"Driver: {self.name}, Vehical: {self.vehical}, Rating: {self._rating}, Available: {self.is_available}"
#Toggle drivers avaibaility status
  def toggle_availability(self): # Indented to be part of the class
    self.is_available = not self.is_available #not True->False
```

##Step 2. Introducing Rider Class

```python
class Rider:
  def __init__(self,name,location):
    self.name = name
    self.location = location

  #Method to request a ride
  def request_ride(self,destination):
    self.destination = destination
    print(f"{self.name} requested a ride from {self.location} to {self.destination}")

  #Polymorphism __str__ repressentation for rider class
  def __str__(self):
    return f"Rider: {self.name}, Location: {self.location}"
```

##Step 3. Rider Sharing APP

```python
class RiderSharingApp:
  def __init__(self):
    self.drivers =[] #Stores all the drivers
    self.riders =[] #Stores all the riders

  #Add driver to the list
  def add_driver(self,driver):
    self.drivers.append(driver)
    print(f"Driver {driver.name} added to the system")

  #Add rider to the list
  def add_rider(self,rider):
    self.riders.append(rider)
    print(f"Rider {rider.name} added to the system")

  #Abstraction - User calls this method to find driver
  #They donot worry how system looks for it
  def find_avaible_driver(self):
    for driver in self.drivers:
      if driver.is_available:
        return driver
    return None

  #Book a ride for a rider for certain destination
  def book_ride(self,rider_name,destination):
    #Find the rider by name
    rider = next((r for r in self.riders if r.name == rider_name), None)
    if not rider:
      print("Rider not found")
      return


    # Rider request a ride
    rider.request_ride(destination)

    #Find a avaibile driver
    driver = self.find_avaible_driver()
    if driver:
      driver.toggle_availability() #Mark driver as busy
      print(f"Ride booked! {driver.name} is on the way to {destination}")
    else:
      print("No available drivers at the moment")



  #After completing rider driver is feered up
  def complete_ride(self,driver_name):
    driver = next((d for d in self.drivers if d.name == driver_name),None)
    if driver and not driver.is_available:
      driver.toggle_availability()
      print(f"{driver.name} has completed the ride. Now avaible for next ride")
    else:
      print("Driver not found")
```

##Step 4. Final Demonstration of the System ----

```python
aap = RiderSharingApp()

## Adding drivers
print("Adding drivers")
driver1 = Driver("John","Sedan")
driver2 = Driver("Alice","SUV")

aap.add_driver(driver1)
aap.add_driver(driver2)


## Adding riders
print("\nAdding riders")
rider1 = Rider("Bob","Downtown")
rider2 = Rider("Eve","Uptown")

aap.add_rider(rider1)
aap.add_rider(rider2)

print("\n Booking rides:")
aap.book_ride("Bob","Uptown")

print("\n Driver Status:")
print(driver1) #Polymorphism via __str__ in driver


print("\n Completing the Ride:")
aap.complete_ride("John")

print("\n Book another ride:")
aap.book_ride("Eve","Downtown")

print("\n Driver Status:")
print(driver2)

```
